```python
import tensorflow as tf
import numpy as np
```

```python
celsius=np.array([-40, -10, 0, 8, 15, 22, 38],dtype=float)
fahrenheit=np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

```python
'''Las capas densas son las capas que tienen conexiones desde cada neurona hacia
Todas las neuronas de la sigiente capa. Las unidades corresponden a la cantidad
de neuronas de la capa de salida y la variable input_shape acepta el numero de
neuronas de la entrada'''
capa = tf.keras.layers.Dense(units=1, input_shape=[1])
#modelo secuencial
modelo = tf.keras.Sequential([capa])
```

```python
''' El siguiente paso es compilar el modelo, lo que lo prepara para
ser entrenado. Tengo que decirle al modelo algunas propiedades para
indicarle como quiero que aprenda mejor. En este caso sólo se utilizarán
dos propiedades: el optimizador y la función de pérdida.
El optimizador le dice a la red como ajustar los pesos de manera eficiente
 para que aprenda y no desaprenda. El 0.1 indica la tasa de aprendizaje, lo que
 le indica a la red que tanto ajustar los pesos y sesgos. La función de pérdida
 '''
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

```python
historial = modelo.fit(celsius, fahrenheit, epochs=500, verbose=False)
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pérdida")
plt.plot(historial.history["loss"])
```

```python
print('Realizando una prediccion.')
input_data = np.array([[100]])
resultado = modelo.predict(input_data)
print('El resultado es: '+ str(resultado) + ' fahrenheit')
```

# ¿Cómo funcionaría el modelo si agregamos dos capas intermedias con 3 neuronas a la red?

image.png


```python
celsius=np.array([-40, -10, 0, 8, 15, 22, 38],dtype=float)
fahrenheit=np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

```python
oculta1 = tf.keras.layers.Dense(units=3, input_shape=[1])
oculta2 = tf.keras.layers.Dense(units=3)
salida = tf.keras.layers.Dense(units=1)
modelo = tf.keras.Sequential([oculta1, oculta2, salida])
```

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.01),
    loss='mean_squared_error'
)
```

```python
historial=modelo.fit(celsius, fahrenheit, epochs=500, verbose=False)
print("¡Modelo entrenado!")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pérdida")
plt.plot(historial.history["loss"])
```

```python
print('Realizando una prediccion.')
input_data = np.array([[100]])
resultado = modelo.predict(input_data)
print('El resultado es: '+ str(resultado) + ' fahrenheit')
```
